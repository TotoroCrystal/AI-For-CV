```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def median_blur(img, kernel, padding_way): # original image, kernel size, padding way: zero or replica
    H = img.shape[0]
    W = img.shape[1]
    M, N = kernel.shape
    adds = kernel//2
    padding_left = (N - 1) // 2
    padding_up = (M - 1) // 2
    
    img_padding = np.zeros((H + M - 1, W + N - 1))
    
    if padding_way == "ZERO":
        img_padding[padding_up : H + padding_up, padding_left : W + padding_left] = img
        
    if padding_way == "REPLICA":
        img_padding[padding_up : H + padding_up, padding_left : W + padding_left] = img
        
        img_padding[ : padding_up, padding_left : W + padding_left] = img[0, :]         # up
        img_padding[H + padding_up : , padding_left : W + padding_left] = img[-1, :]    # bottom
        img_padding[ : , : padding_left] = img_padding[ : , padding_left : padding_left + 1]               # left
        img_padding[ : , W + padding_left : ] = img_padding[ :, W + padding_left - 1 : W + padding_left]   # right
        
    img_padding = img_padding.astype('uint8')
    out_img = img_padding

    # filter the paddding img
    for i in range(H):
        for j in range(W):
            vec = []
            for ki in range(i, i+M):
                for kj in range(j, j+N):
                    vec.append(img_padding[ki, kj])
            # find median pixel value
            vec.sort()
            out_img[i, j] = vec[(len(vec)+1)//2]
    # convert to uint8
    out_img = out_img.astype('uint8')
    
    return out_img
```

```python
kernel = np.zeros((5,5))
img = cv2.imread('salt_noise.jpg')
img = img.astype('uint8')
b, g, r = cv2.split(img)

b_out_zero = median_blur(b, kernel, "ZERO")
g_out_zero = median_blur(g, kernel, "ZERO")
r_out_zero = median_blur(r, kernel, "ZERO")

b_out_replica = median_blur(b, kernel, "REPLICA")
g_out_replica = median_blur(g, kernel, "REPLICA")
r_out_replica = median_blur(r, kernel, "REPLICA")

merged_zero = cv2.merge([b_out_zero, g_out_zero, r_out_zero])
merged_zero = merged_zero.astype('uint8')
merged_replica = cv2.merge([b_out_replica, g_out_replica, r_out_replica])
merged_replica = merged_replica.astype('uint8')
plt.figure(figsize = (16, 9))
plt.subplot(1,3,1)
plt.title('salt_noise')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(1,3,2)
plt.title('padding_zero')
plt.imshow(cv2.cvtColor(merged_zero, cv2.COLOR_BGR2RGB))

plt.subplot(1,3,3)
plt.title('padding_replica')
plt.imshow(cv2.cvtColor(merged_replica, cv2.COLOR_BGR2RGB))
plt.show()
```
