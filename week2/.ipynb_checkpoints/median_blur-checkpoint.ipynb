```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def median_blur(img, kernel, padding_way): # original image, kernel size, padding way: zero or replica
    H = img.shape[0]
    W = img.shape[1]
    M, N = kernel.shape
    adds = kernel//2
    padding_left = (N - 1) // 2
    padding_up = (M - 1) // 2
    
    img_padding = np.zeros((H + M - 1, W + N - 1))
    
    if padding_way == "ZERO":
        img_padding[padding_up : H + padding_up, padding_left : W + padding_left] = img
        
    if padding_way == "REPLICA":
        img_padding[padding_up : H + padding_up, padding_left : W + padding_left] = img
        
        img_padding[ : padding_up, padding_left : W + padding_left] = img[0, :]         # up
        img_padding[H + padding_up : , padding_left : W + padding_left] = img[-1, :]    # bottom
        img_padding[ : , : padding_left] = img_padding[ : , padding_left : padding_left + 1]               # left
        img_padding[ : , W + padding_left : ] = img_padding[ :, W + padding_left - 1 : W + padding_left]   # right
        
    img_padding = img_padding.astype('uint8')
    out_img = img_padding

    # filter the paddding img
    for i in range(H):
        for j in range(W):
            vec = []
            for ki in range(i, i+M):
                for kj in range(j, j+N):
                    vec.append(img_padding[ki, kj])
            # find median pixel value
            vec.sort()
            out_img[i, j] = vec[(len(vec)+1)//2]
    # convert to uint8
    out_img = out_img.astype('uint8')
    
    return out_img
```

```python
kernel = np.zeros((5,5))
img = cv2.imread('salt_noise.jpg')
img = img.astype('uint8')
b, g, r = cv2.split(img)

b_out_zero = median_blur(b, kernel, "ZERO")
g_out_zero = median_blur(g, kernel, "ZERO")
r_out_zero = median_blur(r, kernel, "ZERO")

b_out_replica = median_blur(b, kernel, "REPLICA")
g_out_replica = median_blur(g, kernel, "REPLICA")
r_out_replica = median_blur(r, kernel, "REPLICA")

merged_zero = cv2.merge([b_out_zero, g_out_zero, r_out_zero])
merged_zero = merged_zero.astype('uint8')
merged_replica = cv2.merge([b_out_replica, g_out_replica, r_out_replica])
merged_replica = merged_replica.astype('uint8')
plt.figure(figsize = (16, 9))
plt.subplot(1,3,1)
plt.title('salt_noise')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(1,3,2)
plt.title('padding_zero')
plt.imshow(cv2.cvtColor(merged_zero, cv2.COLOR_BGR2RGB))

plt.subplot(1,3,3)
plt.title('padding_replica')
plt.imshow(cv2.cvtColor(merged_replica, cv2.COLOR_BGR2RGB))
plt.show()
```

### 直方图实现快速中值滤波

传统中值滤波的算法一般是在图像中建立窗口，然后对窗口内的所有像素值进行排序，选择排序后的中间值作为窗口中心像素滤波后的值。

黄色区域+中间粉色区域是第一个像素为中心建立的滤波窗口，粉色区域+右边蓝色区域为同一行第二个像素为中心建立的滤波窗口。传统做法对每一个窗口内所有像素排序，那么中间粉色区域的像素就会被重复的做排序运算，存在大量冗余计算。如果窗口移动一个像素的时候只用考虑去除左边一列内（黄色区域）的像素，并且加上右边一列（蓝色区域）的像素，运算会大大简化。这样的操作可以使用直方图来实现。

1.读取图像I，并且设定滤波窗口大小(winX*winY)，一般winX=winY，奇数。

2.设定中值滤波直方图中的阈值，Thresh=(winX*winY)/2 +1;

3.如果要考虑边界情况，可以先对原图像进行扩展，左、右边界分别扩展winX/2个像素，上下边界分别扩展winY/2个像素。

4.逐行遍历图像像素，以第一行为例：先取第一行第一个要处理的像素（窗口中心像素），建立滤波窗口，提取窗口内所有像素值（N=winX*winY个），获取N个像素的直方图Hist。从左到右累加直方图中每个灰度层级下像素点个数，记为sumCnt，直到sumCnt>=Thresh，这时的灰度值就是当前窗口内所有像素值的中值MediaValue。将MediaValue值赋值给窗口中心像素，表明第一个像素中值滤波完成。

5.此时窗口需要向右移动一个像素，开始滤波第二个像素，并且更新直方图。以第二个像素为窗口中心建立滤波窗口，从前一个窗口的灰度直方图Hist中减去窗口中最左侧的一列像素值的灰度个数，然后加上窗口最右侧一列像素值的灰度个数。完成直方图的更新。

6.直方图更新后，sumCnt值有三种变化可能：（1）减小（2）维持不变（3）增大。这三种情况与减去与加入的像素值灰度有关。此时为了求得新的中值，需要不断调整sumCnt与Thresh之间的关系。

（1）如果sumCnt值小于Thresh：说明中值在直方图当前灰度层级的右边，sumCnt就依次向右加上一个灰度层级中灰度值个数，直到满足sumCnt>=Thresh为止。记录此时的灰度层级代表的灰度值，更新MediaValue，作为第二个像素的滤波后的值。

（2）维持不变：说明MediaValue值不变，直接作为第二个像素滤波后的值。

（3）如果sumCnt值大于Thresh：说明中值在直方图当前灰度层级的左边，sumCnt就依次向左减去一个灰度层级中灰度值个数，直到满足sumCnt<=Thresh为止。记录此时的灰度层级代表的灰度值，更新MediaValue值，作为第二个像素的滤波后的值。

7.窗口逐行依次滑动，求得整幅图像的中值滤波结果。
